```python
import numpy as np
import pandas as pd
import tensorflow as tf
from keras import Sequential
from keras.layers import *
```

```python
train_data = np.loadtxt("/kaggle/input/digit-recognizer/train.csv", delimiter = ',', skiprows=1)
test_data = np.loadtxt("/kaggle/input/digit-recognizer/test.csv", delimiter = ',', skiprows=1)
```

```python
train_data.shape
```

```python
train_size = int(0.7 * train_data.shape[0])
validation_size = int(0.2 * train_data.shape[0])
test_size = 0.1

train_data_sep = train_data[:train_size]
X_train = train_data_sep[:, 1:].reshape(-1,28,28,1)
y_train = train_data_sep[:, 0:1]

validation_data_sep = train_data[train_size:train_size + validation_size]
X_validation = validation_data_sep[:, 1:].reshape(-1,28,28,1)
y_validation = validation_data_sep[:, 0:1]

test_data_sep = train_data[train_size + validation_size:]
X_test = test_data_sep[:, 1:].reshape(-1,28,28,1)
y_test = test_data_sep[:, 0:1]

print(f'Train data size: {X_train.shape[0]} \nPrecent of all data: {X_train.shape[0] / train_data.shape[0] * 100:.2f} %\n')
print(f'Validation data size: {X_validation.shape[0]} \nPrecent of all data: {X_validation.shape[0] / train_data.shape[0] * 100:.2f} %\n')
print(f'Test data size: {X_test.shape[0]} \nPrecent of all data: {X_test.shape[0] / train_data.shape[0]  * 100:.2f} %')
```

```python
input_size = X_train.shape[1]
output_size = 10
hidden_layer = 100

model = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', padding='same', input_shape = (28,28,1)),
    Conv2D(32, kernel_size=(3, 3), activation='relu' ),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(0.25),
    
    Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same'),
    Conv2D(64, kernel_size=(3, 3), activation='relu' ),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(0.25),
    
    Conv2D(128, kernel_size=(3, 3), activation='relu', padding='same' ),
    Conv2D(128, kernel_size=(3, 3), activation='relu' ),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(0.25),
    
    
    Flatten(),
    
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(output_size, activation = "softmax")
])

model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = 'accuracy')

num_epochs = 100

model.fit(X_train, y_train, epochs = num_epochs,
          validation_data = (X_validation, y_validation),
          verbose = 2, batch_size = 128)
```

```python
test_loss, test_accuracy =  model.evaluate(X_test, y_test, verbose = 0)

print(f'Test loss: {test_loss * 100:.2f} %')
print(f'Test accuracy: {test_accuracy * 100:.2f} %')
```

```python
output_size = 10
hidden_layer = 100

model_last = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', padding='same', input_shape = (28,28,1)),
    Conv2D(32, kernel_size=(3, 3), activation='relu' ),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(0.25),
    
    Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same'),
    Conv2D(64, kernel_size=(3, 3), activation='relu' ),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(0.25),
    
    Conv2D(128, kernel_size=(3, 3), activation='relu', padding='same' ),
    Conv2D(128, kernel_size=(3, 3), activation='relu' ),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(0.25),
    
    
    Flatten(),
    
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(output_size, activation = "softmax")
])

model_last.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = 'accuracy')

num_epochs = 50

model_last.fit(train_data[:,1:].reshape(-1,28,28,1), 
          train_data[:,0:1], 
          epochs = num_epochs,
          validation_split = 0.1,
          verbose = 2, 
          batch_size = 128)
```

```python
prediction = model_last.predict(test_data.reshape(-1,28,28,1))
prediction = np.argmax(prediction, axis = 1)
```

```python
submission = pd.read_csv("/kaggle/input/digit-recognizer/sample_submission.csv", index_col = 0)
submission['Label'] = prediction
```

```python
submission.to_csv('sample_submission_last_2.csv')
```
